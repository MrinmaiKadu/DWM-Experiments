```python
#EXP5 - K-means clustering 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic data
n_samples = 300
n_features = 2
n_clusters = 3

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

# Create a K-Means model
kmeans = KMeans(n_clusters=n_clusters)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualize the data and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1,], c='red', marker='x', s=100)
plt.title("K-Means Clustering")
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Create some sample data
X = np.array([[1, 2], [1, 4], [2, 2], [4, 2], [4, 4], [3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('original data')
plt.show()

# Create a KNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier to the data
knn.fit(X, y)

# Make a prediction for a new point
new_point = np.array([[2, 3]])
prediction = knn.predict(new_point)

# Visualize the prediction
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(new_point[0, 0], new_point[0, 1], c='red', marker='x', s=100, label=f'Prediction: {prediction[0]}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('K-Nearest Neighbors Prediction')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyclustering.cluster.kmedoids import kmedoids

X=np.array([[3,4],[4,5],[6,7],[8,9],[9,7]])
#number of clusters =2
k=2
kmedoids_instance=kmedoids(X, initial_index_medoids=np.random.choice(len(X),k, replace=False))

kmedoids_instance.process()
clusters= kmedoids_instance.get_clusters()
medoids=kmedoids_instance.get_medoids()

for i, cluster in enumerate(clusters):
    points=X[cluster]
    plt.scatter(points[:, 0],points[:,0],label=f'cluster {i+1}')


plt.scatter(X[medoids,0],X[medoids,1],c='red',marker='x',s=100,label='Medoids')
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.title("kmedoids Clustering ")
plt.legend()
plt.show()
    
    
```
